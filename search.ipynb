```python
from PIL import Image
from IPython.display import display

def load_image(image_path):
    raw_image = Image.open(image_path).convert('RGB')   

    w,h = raw_image.size
    display(raw_image.resize((w//5,h//5)))
```

```python
from argparse import Namespace
from omegaconf import OmegaConf
```

```python
args = Namespace(
    search_query="a lightbulb",
    search_directory="/net/acadia10a/data/zkhan/winoground/images",
    checkpoint="/net/acadia10a/data/zkhan/ALBEF-checkpoints/ablations/shared_txt_enc_pseudolabels_multimodal_mim_momentum/checkpoint_29.pth",
    text_encoder="bert-base-uncased",
    device="cuda:0"
)
```

```python
load_image(dataset.images[score_matrix_t2i.argmax()])
```
